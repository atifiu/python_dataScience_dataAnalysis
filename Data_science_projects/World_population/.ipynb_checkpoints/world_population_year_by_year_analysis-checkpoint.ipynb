```python
"""Find some of the below stats
1.Year wise total world population
2.Year wise average world population
"""

import pandas as pd
world_pop = pd.read_csv('datasets/world_population.csv')
world_pop.head()
#help(pd.DataFrame.sort_index)
```

```python
world_pop_idx_country_code =  world_pop.loc[:,world_pop.columns != 'year_'].set_index("country_code").sort_index()
world_pop_idx_country_code.head()
```

# World Total Population year wise since 1960

```python
world_pop_total_yearwise = world_pop_idx_country_code.sum(numeric_only = True)
world_pop_total_yearwise = world_pop_total_yearwise.to_frame(name = "Total_population")
world_pop_total_yearwise.sort_values('Total_population', ascending = False, inplace= True )
world_pop_total_yearwise.head(20)

```

```python
world_pop_avg_yearwise = world_pop_idx_country_code.mean(numeric_only = True)
world_pop_avg_yearwise = world_pop_avg_yearwise.to_frame(name = "Avg_population")
world_pop_avg_yearwise.sort_values('Avg_population', ascending = False, inplace= True )
world_pop_avg_yearwise.head(20)
```

```python
columns_list = world_pop_idx_country_code.columns
print(columns_list)
columns_list = columns_list[3:]
print(columns_list)
complete_world_pop_rank = []
for column in columns_list :

    world_pop_rank = world_pop_idx_country_code.nlargest(n = 10, columns = column)
    world_pop_rank["position"] = world_pop_rank["year_1960"].rank(ascending = False)
    world_pop_rank["year"] = column
    world_pop_rank["population"] = world_pop_rank[column]
    complete_world_pop_rank.append(world_pop_rank)
    #world_pop_rank["country_name", "position", "year"]
complete_world_pop_rank = pd.concat(complete_world_pop_rank)
#print(appended_data.head())
#world_pop_rank[["country_name", "position", "year", "population"]].head()
complete_world_pop_rank = complete_world_pop_rank[["country_name", "position", "year", "population"]]
```

```python
complete_world_pop_rank[complete_world_pop_rank["year"] == "year_2019"]
```

```python
import matplotlib.pyplot as plt
#complete_world_pop_rank.plot.bar(x = "year", title = "Rank in terms of population since 1960")
complete_world_pop_vs_year = complete_world_pop_rank[["country_name", "year", "population"]]
complete_world_pop_vs_year.plot( y="population", x = "year", title = "Rank in terms of population since 1960", kind='bar', stacked=True, color=['green', 'red', 'blue', 'cyan', 'yellow', 'orange', 'purple'])
#kind='bar', stacked=True
plt.show()
```

```python
world_pop_idx_country_code.head()
column_list = world_pop_idx_country_code.columns[3:]
#print(column_list)
world_pop_pivot_yearwise = world_pop_idx_country_code.pivot_table(values=column_list, index = "country_code", columns = ["country_name"]).min()

world_pop_pivot_yearwise = world_pop_pivot_yearwise.to_frame(name = "Total_population").reset_index(inplace = False)
#world_pop_pivot_yearwise = world_pop_pivot_yearwise.reset_index(drop = True).rename(columns={world_pop_pivot_yearwise.index.name:'bar'})
#print(world_pop_pivot_yearwise.tail(20))
world_pop_pivot_yearwise.rename(columns = {'level_0':'year'}, inplace = True)
#print(world_pop_pivot_yearwise.tail(20))
#print(type(world_pop_pivot_yearwise))
world_pop_pivot_yearwise.info()
#world_pop_pivot_yearwise[world_pop_pivot_yearwise['country_name'] == 'India']
#world_pop_pivot_yearwise['country_name'].unique()
world_pop_pivot_yearwise.groupby(["year"])["Total_population"].agg(['max', 'min', 'sum'])
world_pop_pivot_yearwise.groupby(["country_name"])["Total_population"].agg(['max', 'min', 'sum'])
```

```python

india = world_pop_pivot_yearwise[world_pop_pivot_yearwise['country_name'] == 'India']
china = world_pop_pivot_yearwise.query('country_name == "China"')
plt.figure(figsize=(30, 20))
plt.plot(india["year"],india["Total_population"],  label='India')
plt.plot(china["year"],china["Total_population"],  label='China')
plt.xticks(np.arange(0, 100, step=10))  # Set label locations.
plt.legend(fontsize = 50)
plt.show()
```

```python
india.plot(x= "year",y = "Total_population",  label='India', kind = "barh")
plt.xticks(np.arange(0, 100, step=10))  # Set label locations.
plt.legend(fontsize = 10)
china.plot(x= "year",y = "Total_population",  label='China',  kind = "barh")
plt.xticks(np.arange(0, 100, step=10))  # Set label locations.
plt.legend(fontsize = 10)
plt.show()
```

```python
world_pop_pivot_yearwise_tmp = world_pop_pivot_yearwise.groupby('year')["Total_population"].agg(['idxmax', 'idxmin'])
new_list = []
#print(world_pop_pivot_yearwise_tmp.head())
for  index , row in world_pop_pivot_yearwise_tmp.iterrows():
    #print("Index :" + index)
    #print("Value of row is " + str(row["idxmax"])+" "+ str(row["idxmin"]) )
    #print(world_pop_pivot_yearwise.at[row["idxmin"],"country_name"])
    #print(world_pop_pivot_yearwise.at[row["idxmin"],"Total_population"])
    #print(world_pop_pivot_yearwise.iloc[[row["idxmax"]]])
    #print(world_pop_pivot_yearwise.iloc[[row["idxmin"]]])
    #print(world_pop_pivot_yearwise)
    new_list.append([index, world_pop_pivot_yearwise.at[row["idxmin"],"country_name"],world_pop_pivot_yearwise.at[row["idxmin"],"Total_population"],
               world_pop_pivot_yearwise.at[row["idxmax"],"country_name"],world_pop_pivot_yearwise.at[row["idxmax"],"Total_population"]])
#print(new_list)
new_list = np.array(new_list)

print(type(new_list))
world_pop_max_min_country = pd.DataFrame(new_list, columns = ["year", "country_min_pop", "min_population", "country_max_pop", "max_population"])
world_pop_max_min_country.head()
```
